```python
# https://www.kaggle.com/datasets/abcsds/pokemon/data
# 포켓몬 게임 데이터셋 EDA

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 데이터 이해
# : 각 포켓몬의 ID
# 이름 : 각 포켓몬의 이름
# 유형 1 : 각 포켓몬에는 유형이 있으며, 이는 공격에 대한 약점/저항을 결정합니다.
# 유형 2 : 일부 포켓몬은 이중 유형이며 2개를 가지고 있습니다.
# 전체 : 이 이후에 나오는 모든 통계의 합계, 포켓몬이 얼마나 강한지에 대한 일반적인 가이드
# HP : 체력 또는 건강은 포켓몬이 기절하기 전에 견딜 수 있는 피해량을 정의합니다.
# Attack : 일반 공격의 기본 수정자(예: 스크래치, 펀치)
# 방어 : 일반 공격에 대한 기본 피해 저항
# SP Atk : 특수 공격, 특수 공격의 기본 수정자(예: 화염 폭발, 버블빔)
# SP Def : 특수 공격에 대한 기본 피해 저항
# 속도 : 각 라운드마다 어떤 포켓몬이 먼저 공격할지 결정합니다.
```

```python
# 데이터 로드

original_data = pd.read_csv('Pokemon.csv')
pokemon = original_data.copy()
```

```python
print(pokemon.shape)
pokemon.head()
```

```python
stats = ["HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"]

# 각 능력치의 합을 계산합니다.
sum_stats = pokemon[stats].sum(axis=1)

# 계산된 합과 'Total' 칼럼의 값이 일치하는지 확인합니다.
check_total = sum_stats == pokemon["Total"]

# 일치하지 않는 경우의 수를 계산합니다.
num_mismatches = (~check_total).sum()

print(f"총 {num_mismatches}개의 포켓몬에서 능력치의 합이 'Total' 값과 일치하지 않습니다.")
print(sum_stats)


print((pokemon[stats].sum(axis=1) !=  pokemon["Total"]).sum())
```
