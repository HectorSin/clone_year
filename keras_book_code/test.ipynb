```python
import pandas as pd
import keras_tuner
import numpy as np
import plotly.express as px
from sklearn.preprocessing import LabelEncoder
```

```python
# xlsx 파일을 불러옴
slp_raw = pd.read_excel('sleeprecord.xlsx')

# 범주형 데이터 정수형 인코딩
label_encoder = LabelEncoder()
slp_raw['오전 집중'] = label_encoder.fit_transform(slp_raw['오전 집중'])

slp_raw.head()
```

```python
# 특정 열 선택하여 병합 데이터 프레임 생성
selected_columns = ['자기전 누워있던 시간', '오전 집중', '총 수면 시간']  # 필요에 따라 열 이름 조정
df_selected = slp_raw[selected_columns]
```

```python
# Parallel coordinates plot 생성
fig = px.parallel_coordinates(df_selected, color="오전 집중",
                              labels={"오전 집중": "오전 집중", 
                                      "자기전 누워있던 시간": "자기전 누워있던 시간",
                                      "총 수면 시간": "총 수면 시간"},
                              color_continuous_scale=px.colors.diverging.Tealrose,
                              color_continuous_midpoint=2)
```

```python
fig.show()
```
